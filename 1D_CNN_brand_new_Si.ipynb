```python
import pandas as pd
import numpy as np
import time
from keras.models import Sequential
from keras.layers import Dense, Conv1D, MaxPooling1D, GlobalMaxPooling1D
from sklearn.model_selection import train_test_split
```

```python
df_X = pd.read_pickle(r"C:\Users\Hong\Downloads\MediaPipe_Testing\MediaPipePoseEstimation\Deep_Learning\Y_array.pkl")
df_Y = pd.read_pickle(r"C:\Users\Hong\Downloads\MediaPipe_Testing\MediaPipePoseEstimation\Deep_Learning\Group.pkl")
```

```python
# Convert the lists to numpy arrays for compatibility with Keras
X = np.array(df_X)
Y = np.array(df_Y)
```

```python
# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Initialize the model
#model = Sequential()

# Add the first convolution layer
#model.add(Conv1D(filters=200, kernel_size=10, activation='tanh', input_shape=(33, 847)))
#model.add(Conv1D(filters=200, kernel_size=10, activation='tanh'))
#model.add(MaxPooling1D(pool_size=2))

# Add the second convolution layer
#model.add(Conv1D(filters=200, kernel_size=10, activation='tanh'))
#model.add(Conv1D(filters=200, kernel_size=10, activation='tanh'))
#model.add(MaxPooling1D(pool_size=2))

# Add the third convolution layer
#model.add(Conv1D(filters=200, kernel_size=10, activation='tanh'))
#model.add(Conv1D(filters=200, kernel_size=10, activation='tanh'))

# Add a GlobalMaxPooling1D layer
#model.add(GlobalMaxPooling1D())

# Add a Dense layer with the appropriate number of units and activation function
# For binary classification
#model.add(Dense(3, activation='sigmoid'))

#model.summary()

# Initialize the model
model = Sequential()

# Add the first convolution layer
model.add(Conv1D(filters=200, kernel_size=3, activation='tanh', input_shape=(33, 847)))
model.add(Conv1D(filters=200, kernel_size=3, activation='tanh'))
model.add(MaxPooling1D(pool_size=2))

# Add the second convolution layer
model.add(Conv1D(filters=200, kernel_size=3, activation='tanh'))
model.add(Conv1D(filters=200, kernel_size=3, activation='tanh'))
model.add(MaxPooling1D(pool_size=2))

# Add the third convolution layer
model.add(Conv1D(filters=200, kernel_size=3, activation='tanh'))
model.add(Conv1D(filters=200, kernel_size=3, activation='tanh'))

# Add a GlobalMaxPooling1D layer
model.add(GlobalMaxPooling1D())

# Add a Dense layer with the appropriate number of units and activation function
# For binary classification
model.add(Dense(3, activation='sigmoid'))

model.summary()
```

```python
#model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
model.compile(loss='sparse_categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
start_time = time.time()  # Start the timer
```

```python
# Train the model
model.fit(X_train, Y_train, batch_size=2, epochs=1000)
```

```python
end_time = time.time()  # End the timer
```

```python
print(f"The processing time of the SVM is {end_time - start_time} seconds.")  # Print the processing time
```

```python
# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, Y_test, batch_size=2)
```

```python
# Print the test loss and test accuracy
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
# Make predictions on the test set
Y_pred = model.predict(X_test)
```

```python
# Convert predictions to class labels
Y_pred_classes = np.argmax(Y_pred, axis=1)
```

```python
from sklearn.metrics import classification_report
```

```python
# Print the classification report
print(classification_report(Y_test, Y_pred_classes))
```
